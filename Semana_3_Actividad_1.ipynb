```python
input = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'

import pandas as pd
import numpy as np
df = pd.read_csv(input)
df
```

```python
df1 = df.copy() #se realiza copia del archivo original
df1.isnull().values.any() #se verifica si hay datos faltantes R/ si
```

```python
df1.isnull().any() # se evidencia que desde la columna X2 hasta la Y hay campos vacios
```

```python
df1.describe() #metricas estadisticas de los datos
```

```python
df1.isna().sum() #con esta informacion no se eliminaran los campos y se buscara reemplazarlos, esto en la medida que los datos faltantes no son representativos dentro de la muestra
```

```python
edad = df1.X5.mean()  #para el caso de la edad en los datos faltantes se reemplazar por el promedio

df1['X2'].fillna(value = df1.X2.median(), inplace = True)
df1['X3'].fillna(value = 4, inplace = True) #se reemplazan los 2 valores por la opcion 4 =others
df1['X4'].fillna(value = 3, inplace = True) #se reemplazan los 2 valores por la opcion 3 =others
df1['X5'].fillna(value = edad, inplace = True)

# en los campos de historial de pagos atrasados y en la medida que los datos faltantes no son representativos se reemplazan por la moda de cada serie
df1['X6'].fillna(value = df1.X6.mode()[0], inplace = True)
df1['X7'].fillna(value = df1.X7.mode()[0], inplace = True)
df1['X8'].fillna(value = df1.X8.mode()[0], inplace = True)
df1['X9'].fillna(value = df1.X9.mode()[0], inplace = True)
df1['X10'].fillna(value = df1.X10.mode()[0], inplace = True)
df1['X11'].fillna(value = df1.X11.mode()[0], inplace = True)

# en los campos de estado de cuenta y monto pagado anterior, ya que en promedio representan aprox el 0.05% de la muestra se decide dejar en 0 los na, de esta forma las metricas de cada campo no se ven afectadas

df1['X12'].fillna(value = 0, inplace = True)
df1['X13'].fillna(value = 0, inplace = True)
df1['X14'].fillna(value = 0, inplace = True)
df1['X15'].fillna(value = 0, inplace = True)
df1['X16'].fillna(value = 0, inplace = True)
df1['X17'].fillna(value = 0, inplace = True)
df1['X18'].fillna(value = 0, inplace = True)
df1['X19'].fillna(value = 0, inplace = True)
df1['X20'].fillna(value = 0, inplace = True)
df1['X21'].fillna(value = 0, inplace = True)
df1['X22'].fillna(value = 0, inplace = True)
df1['X23'].fillna(value = 0, inplace = True)


df1.isna().sum()
df1.describe()

```

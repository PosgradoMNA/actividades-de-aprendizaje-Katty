```python
import pandas as pd
import matplotlib.pylab as plt
from matplotlib import pyplot
import numpy as np

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
cars = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data', names = headers)
cars.head()
```

```python
cars.replace("?", np.nan, inplace = True) #se reemplzan los ? por NaN
cars.head()
```

```python
missing_data = cars.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")  
```

```python
# normalizacion

normalized_losses_avg = cars['normalized-losses'].astype(float).mean()
cars['normalized-losses'].replace(np.nan, normalized_losses_avg, inplace = True)

bore_avg = cars['bore'].astype(float).mean()
cars['bore'].replace(np.nan, bore_avg, inplace = True)

stroke_avg = cars['stroke'].astype(float).mean()
cars['stroke'].replace(np.nan, stroke_avg, inplace = True)

horsepower_avg = cars['horsepower'].astype(float).mean()
cars['horsepower'].replace(np.nan, horsepower_avg, inplace = True)

peak_rpm_avg = cars['peak-rpm'].astype(float).mean()
cars['peak-rpm'].replace(np.nan, peak_rpm_avg, inplace = True)
```

```python
cars['num-of-doors'].value_counts().idxmax()
```

```python
cars["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
cars.dropna(subset=["price"], axis=0, inplace=True)
cars.reset_index(drop=True, inplace=True)
cars.head()
```

```python
cars.dtypes
```

```python
cars[["bore", "stroke"]] = cars[["bore", "stroke"]].astype("float")
cars[["normalized-losses"]] = cars[["normalized-losses"]].astype("int")
cars[["price"]] = cars[["price"]].astype("float")
cars[["horsepower", "peak-rpm"]] = cars[["horsepower", "peak-rpm"]].astype("float")
cars.dtypes
```

#Estandarizacion

Convertir todo a un solo formato

```python
cars['city-L/100km'] = 235/cars["city-mpg"]  # se transforma a una unidad de medida
```

#Normalizacion



```python
cars['length'] = cars['length']/cars['length'].max()
cars['width'] = cars['width']/cars['width'].max()
cars['height'] = cars['height']/cars['height'].max()
```

```python
cars['horsepower'] = cars['horsepower'].astype(int, copy=True)
%matplotlib inline
plt.hist(cars["horsepower"])

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```
